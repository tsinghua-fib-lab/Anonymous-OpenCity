## 绘制原始数据指标分布图

### 1 Seg index
    Beijing, Newyork, San Francisco

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

city = ["Beijing", "Newyork", "San_Francisco"]#,  "London", "Paris", "Sydney"

dat = []
for c in city:
    df = pd.read_csv(f"dataForPlot/{c}_CBG_Segregation_z_n.csv")
    dat.append(df['seg5'].to_numpy())


fig, ax = plt.subplots()

for c, d in zip(city,dat):
    # 为每个数据数组绘制直方图
    ax.hist(d, bins=50, alpha=0.6, label=c)
    

# 添加图例
ax.legend()

# 设置标题和轴标签
ax.set_title('Distribution of Segregation Index')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')

# 显示图表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

city = [ "Newyork-normal", 'Newyork-even', 'San_Francisco-normal', 'San_Francisco-even']#"Beijing",, "San_Francisco",  "London", "Paris", "Sydney"

dat = []
for c in city:

    df = pd.read_csv(f"dataForPlot/{c.split('-')[0]}_genData_CBG_Segregation_{c.split('-')[1]}.csv")
    na = df['seg5'].to_numpy()
    dat.append(na)
    print(f"{c}: 地点数量：{len(na)}, seg均值 {np.mean(na)}")


fig, axes  = plt.subplots(1, 2, figsize=(16, 8), sharex=True)

for  ax, c, d, color in zip([axes[0],axes[0],axes[1],axes[1]], ["New_York-Original", "New_York-Even", "San_Francisco-Original", "San_Francisco-Even"], dat, ['red', 'yellow', 'green', 'blue']):
    # 为每个数据数组绘制直方图
    
    ax.hist(d, bins=30, alpha=0.6, label=c, color=color)
    

    # 添加图例
    ax.legend(fontsize=20)
    ax.set_xlim(0, 1.05)
    ax.set_ylim(0, 100)


tick_label_fontsize = 18
for ax in axes:
    ax.tick_params(axis='both', labelsize=tick_label_fontsize)

axes[1].set_yticklabels([])
# axes[2].set_yticklabels([])
# axes[3].set_yticklabels([])

# # 设置标题和轴标签
# ax.set_title('Distribution of Segregation Index')
# axes[0].set_xlabel('Income Segregation Index')
# axes[1].set_xlabel('Income Segregation Index')
# axes[2].set_xlabel('Income Segregation Index')
# axes[3].set_xlabel('Income Segregation Index')
axes[0].set_ylabel('Frequency', fontsize=20) 

plt.tight_layout()
fig.text(0.5, -0.02, 'Income Segregation Index', ha='center', fontsize=20)

# 显示图表
plt.show()
```
